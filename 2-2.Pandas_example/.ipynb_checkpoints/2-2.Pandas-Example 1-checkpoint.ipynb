# Example 酒類市場銷售分析

#### 題目: 讀取drinks.csv ，進行銷售分析:
* 啤酒(beer_servings)銷售分析
    * 各洲(continent)啤酒(beer_servings)平均銷售量？ 請依序排列。
    * 請列出前五個國家購買啤酒(beer_servings)最多的國家(country)。

* 紅酒(wine_servings銷售分析
    * 以describe()，列出各洲continent的紅酒的銷售分析(count, mean, std, min, 25%,	50%, 75%,max)
    * 
    
* 烈酒(spirit_servings)銷售分析
    * 各洲continent在烈酒spirit_servings飲料市場中，單一國家的平均銷量和最大銷量
    * 烈酒spirit_servings在各洲continent的總銷售量

```python
import pandas as pd

path = 'drinks.csv'
drinks = pd.read_csv(path) 

drinks.head()
```

* 啤酒(beer_servings)銷售分析
    * 各洲(continent)啤酒(beer_servings)平均銷售量？ 請依序排列。

```python
drinks.groupby('continent').beer_servings.mean()
```

* 啤酒(beer_servings)銷售分析
    * 請列出前五個國家購買啤酒(beer_servings)最多的國家(country)。

```python
beer_top5 = drinks.sort_values(by='beer_servings', ascending = False)
beer_top5[:5]
```

* 紅酒(wine_servings銷售分析
    * 以describe()，列出各洲continent的紅酒的銷售分析(count, mean, std, min, 25%,	50%, 75%,max)

```python
drinks.groupby('continent').wine_servings.describe()
```

* 烈酒(spirit_servings)銷售分析
    * 各洲continent在烈酒spirit_servings飲料市場中，單一國家的平均銷量和最大銷量

```python
drinks.groupby('continent').spirit_servings.agg(['mean', 'max'])
```

* 烈酒(spirit_servings)銷售分析
    * 烈酒spirit_servings在各洲continent的總銷售量

```python
drinks.groupby('continent').spirit_servings.sum()
```

# 探索泰坦尼克災難數據

#### 題目: 讀取titanic.csv，進行美國犯罪分析:
* 將PassengerId設為index
* 船上男女乘客的總人數
* 列出船票票價(Fare)平均價格、最高價格與最低價格
* 列出各個艙等(Cabin)的船票票價(Fare)平均價格
* 列出各個艙等(Cabin)的生還人數(Survived)
* 列出三個票價區間(Fare<20, 20>=Fare<200, Fare>=20)的人數和生還人數(Survived)
* 列出男、女乘客的平均年齡(Age)、最高年齡與最低年齡

* 繪製一個展示男女乘客比例的扇形圖
* 船票的票價，與乘客年齡和性別的散點圖
* 船票價格的直方圖




 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
path = '/Users/jason/Downloads/Python課程/2-2.Pandas/titanic.csv'

titanic = pd.read_csv(path)
titanic.head()
```

* 將PassengerId設為index

```python
titanic.set_index('PassengerId').head()
```

* 船上乘客的總人數

```python
titanic.sum()
```

* 船上男女乘客的總人數

```python
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

print('males =', males)
print('females =', females)
```

* 列出船票票價(Fare)平均價格、最高價格與最低價格

```python
print('平均價格 =', titanic['Fare'].mean())
print('最高價格 =', titanic['Fare'].max())
print('最低價格 =', titanic['Fare'].min())
```

* 列出各個艙等(Cabin)的船票票價(Fare)平均價格

```python
titanic.groupby('Cabin').Fare.mean()
```

* 列出各個艙等(Cabin)的生還人數(Survived)

```python
titanic.groupby('Cabin').Survived.sum()
```

* 列出三個票價區間(Fare<20, 20>=Fare<200, Fare>=20)的人數和生還人數(Survived)

```python
a = titanic.Fare<=200
b = titanic.Fare>20

Passeger_200 = titanic[titanic.Fare>200] 
Passeger_200_num = Passeger_200.sum()
Passeger_200_passenger = (Passeger_200['Sex'] == 'male').sum() + (Passeger_200['Sex'] == 'female').sum()

Passeger_20 = titanic[titanic.Fare<=20] 
Passeger_20_num = Passeger_20.sum()
Passeger_20_passenger = (Passeger_20['Sex'] == 'male').sum() + (Passeger_20['Sex'] == 'female').sum()

Passeger_other = titanic[a & b] 
Passeger_other_num = Passeger_other.sum()
Passeger_other_passenger = (Passeger_other['Sex'] == 'male').sum() + (Passeger_other['Sex'] == 'female').sum()

print('票價 Fare<=20 登船人數 =',Passeger_20_passenger,'生還人數  =', Passeger_20_num.loc['Survived'])
print('票價 20>Fare<=200 登船人數 =',Passeger_other_passenger,'生還人數 =', Passeger_other_num.loc['Survived'])
print('票價 Fare>200 登船人數 =',Passeger_200_passenger,'生還人數 =', Passeger_200_num.loc['Survived'])
```

```python

x =[{'Fare<20':Passeger_20_passenger,'20>=Fare<200':Passeger_other_passenger,'Fare>=200':Passeger_20_passenger},{'Fare<20':Passeger_20_num.loc['Survived'], '20>=Fare<200':Passeger_other_num.loc['Survived'],'Fare>=200':Passeger_200_num.loc['Survived']}]

pd.DataFrame(x, index = ['登船人數', '生還人數'])
```

* 列出男、女乘客的平均年齡(Age)、最高年齡與最低年齡

```python
Age_mean = titanic.groupby('Sex').Age.mean()
Age_max = titanic.groupby('Sex').Age.max()
Age_min = titanic.groupby('Sex').Age.min()

print('乘客的平均年齡 =', Age_mean)
print('\n乘客的最高年齡 =', Age_max)
print('\n乘客的最低年齡 =', Age_min)
```

* 繪製一個展示男女乘客比例的扇形圖

```python

males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

proportions = [males, females]

plt.pie(
    proportions,
    labels = ['Males', 'Females'],
    shadow = False,
    colors = ['blue','red'],
    explode = (0.15 , 0),
    startangle = 90,
    autopct = '%1.1f%%'
    )

plt.axis('equal')
plt.title("Sex Proportion")
plt.tight_layout()
plt.show()
```

* 船票的票價，與乘客年齡和性別的散點圖

```python
# creates the plot using
lm = sns.lmplot(x = 'Age', y = 'Fare', data = titanic, hue = 'Sex', fit_reg=False)
 
# set title
lm.set(title = 'Fare x Age')
 
# get the axes object and tweak it
axes = lm.axes
axes[0,0].set_ylim(-5,)
axes[0,0].set_xlim(-5,85)
```

* 船票價格的直方圖

```python
# sort the values from the top to the least value and slice the first 5 items
df = titanic.Fare.sort_values(ascending = False)
df
 
# create bins interval using numpy
binsVal = np.arange(0,600,10)
binsVal
 
# create the plot
plt.hist(df, bins = binsVal)
 
# Set the title and labels
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Payed Histrogram')
 
# show the plot
plt.show()
```
