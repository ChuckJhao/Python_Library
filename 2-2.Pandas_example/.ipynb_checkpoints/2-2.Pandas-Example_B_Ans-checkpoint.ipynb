# 合歡山氣象分析

#### 題目: 讀取weather.xlsx，進行氣象分析:
* 對於全體數據，計算所有氣候資料(最高氣壓、氣溫、相對濕度...)的平均值
* 將"氣溫(℃)"以最低溫度到最高溫度依序列出整張表
* 列出'降水量(mm)'最高的前五個時間，並列出"相對溼度(%)"做為參考。
* 先將中文"氣溫(℃)"轉為"temperature"，列出大於temperature > 6 的時間


```python
import pandas as pd
import numpy as np

weather = pd.read_excel('weather.xlsx',index_col="觀測時間(day)")

weather.head()
```

* 對於全體數據，計算所有氣候資料(最高氣壓、氣溫、相對濕度...)的平均值

```python
weather.describe()
```

```python
weather.mean()
```

* 將"氣溫(℃)"以最低溫度到最高溫度依序列出整張表

```python
result = weather.sort_values("氣溫(℃)")
result
```

* 列出'降水量(mm)'最高的前五個時間，並列出"相對溼度(%)"做為參考。

```python
rain_top5 = weather.sort_values(by='降水量(mm)', ascending=False)

rain_top5[0:5]
#rain_top5.loc[0:5,['降水量(mm)',"相對溼度(%)"]]
```

```python

weather.loc[0:5,['降水量(mm)',"相對溼度(%)"]]
```

* 先將中文"氣溫(℃)"轉為"temperature"
* 列出大於temperature > 6 的時間

```python
w2 = weather.rename(columns={"氣溫(℃)": "temperature"})
w2[w2.temperature>6] 
```

# 探索風速數據

#### 題目: 讀取wind_data.csv，進行分析:
* 缺失資料統計
* 計算全體數據的風速平均值
* 按照年為頻率，顯示採樣資料
* 新建表單loc_stats，計算每個地區的風速偏移，最小值，最大值，平均值和標準差
* 並將表單loc_stats存入檔案wind_loc_stats.csv



```python
import pandas as pd
import datetime

path = 'C:/Users/tsiot07/Desktop/00. Python_Code/2-2.Pandas_example/wind_data.csv'

wind = pd.read_csv(path)
wind
```

* 缺失資料統計

```python
wind.isnull().sum()
```

* 計算全體數據的風速平均值

```python
wind.mean().mean()
```

```python
wind.describe()
```

```python
* 按照年為頻率，顯示採樣資料
```

```python
wind.query('Mo == 1 and Dy == 1')
```

* 新建表單loc_stats，計算每個地區的風速偏移，最小值，最大值，平均值和標準差

```python
loc_stats = pd.DataFrame()

loc_stats['min'] = wind.min() # min
loc_stats['max'] = wind.max() # max 
loc_stats['mean'] = wind.mean() # mean
loc_stats['std'] = wind.std() # standard deviations

loc_stats[3:]
```

* 並將表單loc_stats存入檔案wind_loc_stats.csv

```python
wind_loc_stats = loc_stats[3:].round(2)
wind_loc_stats
```

```python
wind_loc_stats.reset_index(inplace=True)
wind_loc_stats.to_csv('wind_loc_stats.csv', encoding='utf-8', index=False)
```
