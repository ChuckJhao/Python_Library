# 美國犯罪分析

#### 題目: 讀取crime.xlsx，進行美國犯罪分析:
* 將Year設為index
* 刪除名為Total的列
* 何時是美國歷史上生存最危險(各種犯罪Murder...統計值最高)的年代？
* 美國歷史上犯罪Murder最高的前五年？
* 查看每十年(1961, 1971)的犯罪統計，看是否有改善？
* 美國歷史上犯罪Murder最高的前五年？
 


```python
import pandas as pd
import numpy as np

path = r'C:\Users\tsiot07\Desktop\00. Python_Code\2-2.Pandas_example\crime.xlsx'
crime = pd.read_excel(path)

crime.head()
```

* 將Year設為index

```python
crime = crime.set_index('Year', drop = True)
crime.head()
```

* 刪除名為Total的列

```python
del crime['Total']
crime.head()
```

* 何時是美國歷史上生存最危險(各種犯罪Murder...統計值最高)的年代？

```python
crime.idxmax(0)
```

* 美國歷史上犯罪Murder最高的前五年？

```python
crime_top5 = crime.sort_values(by='Murder', ascending=False)
crime_top5[0:5]
```

* 查看每十年(1961, 1971)的犯罪統計，看是否有改善？

```python
crime.query('Year % 10 == 1')
```

# 2012歐洲杯足球賽資料分析

#### 題目: 讀取Euro2012_stats.csv，進行分析:
* 有多少球隊參與了2012歐洲杯？
* 進球數前五的國家
* 將黃牌和紅牌重新整理在discipline的表單
* 按照先紅卡（由多到少）再黃卡進行排序
* 找到進球數進球超過5的球隊數據
* 找到英格蘭（England），義大利（Italy）和俄羅斯（Russia）的射正率


```python
import pandas as pd

path = r'C:\Users\tsiot07\Desktop\00. Python_Code\2-2.Pandas_example\Euro2012_stats.csv'

euro12 = pd.read_csv(path)
euro12
```

* 有多少球隊參與了2012歐洲杯？

```python
euro12.shape[0]
```

* 進球數前五的國家

```python
euro12.sort_values(by='Goals', ascending=False).head()
```

* 將黃牌和紅牌重新整理在discipline的表單

```python
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
```

* 按照先紅卡（由多到少）再黃卡進行排序

```python
discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending = False)
```

* 找到進球數進球超過5的球隊數據

```python
euro12[euro12.Goals > 5]
```

* 找到英格蘭（England），義大利（Italy）和俄羅斯（Russia）的射正率

```python
euro12.loc[euro12.Team.isin(['England', 'Italy', 'Russia']), ['Team','Shooting Accuracy']]
```
